## Preparation

```python
## importing the file.
df = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
df.head(5)
```

```python
## replacing diagnoses.

df["diagnosis"] = df["diagnosis"].replace({
    "M":"malignant",
    "B":"benign"
})

df.head(5)
```

## Data Analysis

```python
## areas of tumors for malignants and benign tumors.
plt.style.use("ggplot")

data = df.groupby("diagnosis").agg("count")["area_mean"].to_frame().reset_index()
data_x = data["diagnosis"]
data_y = data["area_mean"]

fig, ax = plt.subplots()
ax.bar(data_x,data_y)

for i,v in enumerate(data["area_mean"]):
    ax.text(i,v-20,str(v),ha="center",fontsize=8,color="white")
    
ax.text(1,300,f'Total = {data["area_mean"].sum()}',ha="center",fontstyle="italic")

benign = (df["diagnosis"].loc[df["diagnosis"] == "benign"].count() / len(df["diagnosis"])) * 100
ax.set_title(f'{str(round(benign,2))}% of the observed tumors were benign.',fontsize=12)
ax.set_ylabel("Measurement (cm)",fontsize=10)

plt.show()
```

### Size-related statistics

```python
## Area Mean statistics

data = df[["diagnosis","radius_mean","perimeter_mean","area_mean"]]
data_benign = data.loc[data["diagnosis"] == "benign"]
data_mal = data.loc[data["diagnosis"]=="malignant"]

## plotting benign 

fig,(ax1,ax2,ax3) = plt.subplots(3,1,figsize=(7,15))

ax1.boxplot([data_benign["area_mean"],data_mal["area_mean"]],labels=["Benign","Malignant"])
ax1.set_title("Malignant tumors have significantly larger area that benign tumors.",fontsize=10)
ax1.set_ylabel("Area (cm squared)",fontsize=8)

ax1.text(3,2300,
        f'Benign $\mu$ = {round(data_benign["area_mean"].mean(),2)} cm squared \n Malignant $\mu$ = {round(data_mal["area_mean"].mean(),2)} cm squared',
        fontsize=8,
        ha="center",
       bbox={"facecolor":"white",
            "edgecolor":"black"})

#### Area Mean

data = df[["diagnosis","radius_mean","perimeter_mean","area_mean"]]
data_benign = data.loc[data["diagnosis"] == "benign"]
data_mal = data.loc[data["diagnosis"]=="malignant"]

## plotting benign 

ax2.boxplot([data_benign["radius_mean"],data_mal["radius_mean"]],labels=["Benign","Malignant"])
ax2.set_title("Malignant tumors have significantly larger radius that benign tumors.",fontsize=10)
ax2.set_ylabel("Radius (cm)",fontsize=8)

ax2.text(3,25,
        f'Benign $\mu$ = {round(data_benign["radius_mean"].mean(),2)} cm squared \n Malignant $\mu$ = {round(data_mal["radius_mean"].mean(),2)} cm squared',
        fontsize=8,
        ha="center",
       bbox={"facecolor":"white",
            "edgecolor":"black"})


#### Perimeter

data = df[["diagnosis","radius_mean","perimeter_mean","area_mean"]]
data_benign = data.loc[data["diagnosis"] == "benign"]
data_mal = data.loc[data["diagnosis"]=="malignant"]

## plotting benign 

ax3.boxplot([data_benign["perimeter_mean"],data_mal["perimeter_mean"]],labels=["Benign","Malignant"])
ax3.set_title("Malignant tumors have significantly larger perimeters that benign tumors.",fontsize=10)
ax3.set_ylabel("Perimeter (cm)",fontsize=8)

ax3.text(3,170,
        f'Benign $\mu$ = {round(data_benign["perimeter_mean"].mean(),2)} cm squared \n Malignant $\mu$ = {round(data_mal["perimeter_mean"].mean(),2)} cm squared',
        fontsize=8,
        ha="center",
       bbox={"facecolor":"white",
            "edgecolor":"black"})

fig.suptitle("Malignant tumors are significantly larger than benign ones in at least 3 dimensions.",fontsize=12)

plt.show()
```

```python
## smoothness

data = df[["diagnosis","smoothness_mean"]]
data_ben = data.loc[data["diagnosis"]=="benign"]
data_mal = data.loc[data["diagnosis"]=="malignant"]

fig,ax = plt.subplots()

ax.boxplot([data_ben["smoothness_mean"],data_mal["smoothness_mean"]],labels=["Benign","Malignant"],vert=False)

ax.text(0.09,2.15,f'$\mu$ = {round(data_mal["smoothness_mean"].mean(),4)} cm', fontsize=8)
ax.text(0.08,1.15,f'$\mu$ = {round(data_ben["smoothness_mean"].mean(),4)} cm', fontsize=8)
ax.set_title("There is slight difference between smoothness for malignant and benign tumors.",fontsize=10)
ax.set_xlabel("Smoothness (cm)",fontsize=10)

plt.show()
```

```python
## Concavity and concave points

data = df[["diagnosis","concavity_mean","concave points_mean"]]

fig,ax = plt.subplots()

ax.scatter(data["concavity_mean"],data["concave points_mean"])
corr_coef = data["concavity_mean"].corr(data["concave points_mean"])
ax.text(0.3,0,
        f'Pearson Coefficient = {round(corr_coef, 2)}',
        fontsize=8,
       fontstyle="italic",
       bbox={"facecolor":"white",
            "edgecolor":"black"})

ax.set_title("There is strong correlation between concavity and concave points.")
ax.set_xlabel("Concavity",fontsize=8)
ax.set_ylabel("Concave Points",fontsize=8)

plt.show()
```
